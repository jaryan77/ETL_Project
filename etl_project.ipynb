```python
#IMPORT DEPENDENCIES
import pandas as pd
from sqlalchemy import create_engine
from config import sqlpassword
```

```python
#IMPORT CSV
movies_metadata_file = "Resources/movies_metadata.csv"
movies_metadata_df = pd.read_csv(movies_metadata_file)
movies_metadata_df
```

```python
#LIST COLOUMNS
list(movies_metadata_df)
```

```python
#FILTER OUT COLOUMNS
movies_metadata_cols = [
 'imdb_id',
 'original_language',
 'tagline',
'overview'
]

movies_metadata_final = movies_metadata_df[movies_metadata_cols].copy()

movies_metadata_final
```

```python
#IMPORT CSV
imdb_file = "Resources/imdb.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df
```

```python
#FILTER THEN RENAME COLOUMNS
imdb_cols = [
'imdb_title_id',
'original_title',
'year',
'date_published',
'genre',
'duration',
'country',
'language',
'director',
'writer',
'actors',
'avg_vote',
'votes'
]

imdb_final = imdb_df[imdb_cols].copy()

imdb_final = imdb_final.rename(columns={ 
'imdb_title_id':'imdb_id',
 'original_title':'movie_title',
 'avg_vote': 'imdb_avg_vote',
'votes':'imdb_vote_count',
'duration':'runtime'
})
```

```python
imdb_final
```

```python
#MERGE THE TWO IMDB DATASETS
combined_imdb = pd.merge(imdb_final, movies_metadata_final, on='imdb_id', how="left")
combined_imdb
```

```python
#IMPORT CSV
rottentomatoes_file = "Resources/rottentomatoes.csv"
rottentomatoes_df = pd.read_csv(rottentomatoes_file)
rottentomatoes_df
```

```python
#LIST COLOUMNS
list(rottentomatoes_df)
```

```python
#FILTER THEN RENAME COLOUMNS
rottentomatoes_cols = [
 'movie_title',
 'critics_consensus',
 'content_rating',
 'genres',
'directors',
 'production_company',
 'tomatometer_rating',
 'tomatometer_count',
 'audience_rating',
 'audience_count',
  'runtime'  
]

rottentomatoes_final = rottentomatoes_df[rottentomatoes_cols].copy()


rottentomatoes_final = rottentomatoes_final.rename(columns={ 
    'directors':'director',
 'tomatometer_rating': 'rottentomatoes_critic_rating',
    'tomatometer_count':'rottentomatoes_critic_count',
 'audience_rating': 'rottentomatoes_audience_rating',
'audience_count': 'rottentomatoes_audience_count'
})
```

```python
list(rottentomatoes_final)
```

```python
#CREATE AND SET INDEX
rottentomatoes_final.reset_index(inplace=True)

rottentomatoes_final = rottentomatoes_final.rename(columns={ 
 'index':"id"
})
```

```python
#CREATE AND SET INDEX
combined_imdb.reset_index(inplace=True)

combined_imdb = combined_imdb.rename(columns={ 
 'index':"id"
})
```

```python
#CHECK DATA TYPES
rottentomatoes_final.dtypes
```

```python
#CHECK DATA TYPES
combined_imdb.dtypes
```

```python
#CREATE ENGINE
engine = create_engine(f"postgresql://postgres:{sqlpassword}@localhost:5432/movies_db2")
```

```python
#TEST/CALL ENGINE TABLE NAMES
engine.table_names()
```

```python
#LOAD DATA INTO POSTGRESQL
combined_imdb.to_sql(name='imdb', con=engine, if_exists='append', index=False)
```

```python
#LOAD DATA INTO POSTGRESQL
rottentomatoes_final.to_sql(name='rottentomatoes', con=engine, if_exists='append', index=False)
```
